```python
# Import dependencies
import pandas as pd
```

```python
# Read CSV file
df = pd.read_csv('/Users/meganconnelly/desktop/dataset.csv')
df.head()
```

```python
conversion_criteria = {
    'Marital status': {
        1: 'Single',
        2: 'Married',
        3: 'Widower',
        4: 'Divorced',
        5: 'Facto Union',
        6: 'Legally Seperated'
    },
    'Daytime/evening attendance': {
        0: 'Evening',
        1: 'Daytime'
    },
    'Gender': {
        0: 'Female',
        1: 'Male'
    },
    'Course': {
        1: 'Biofuel Production Technologies',
        2: 'Animation and Multimedia Design',
        3: 'Social Service',
        4: 'Agronomy',
        5: 'Communication Design',
        6: 'Veterinary Nursing',
        7: 'Informatics Engineering',
        8: 'Equiniculture',
        9: 'Management',
        10: 'Social Service',
        11: 'Tourism',
        12: 'Nursing',
        13: 'Oral Hygiene',
        14: 'Advertising and Marketing Management',
        15: 'Journalism and Communication',
        16: 'Basic Education',
        17: 'Management(evening attendance)'
    },
    'Tuition fees up to date': {
        0: 'No',
        1: 'Yes'
    },
    'Scholarship holder': {
        0: 'No',
        1: 'Yes'
    }
}
```

```python
# Convert columns to numeric types
numeric_columns = ['Marital status', 'Daytime/evening attendance', 'Gender', 'Course', 'Tuition fees up to date', 'Scholarship holder']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
# Perform the conversion for each column
for column, criteria in conversion_criteria.items():
    df[column] = df[column].replace(criteria)
    
df.head()
```

```python
df.to_csv('/Users/meganconnelly/desktop/dataset_1.csv', index=False)
```
